```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo de gráficos
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
```

```python
# Cargar el archivo de datos limpios que guardamos
df = pd.read_csv('../datos/datos_limpios.csv')

# Convertir fechas de nuevo a datetime
df['start_date_time'] = pd.to_datetime(df['start_date_time'])
df['end_date_time'] = pd.to_datetime(df['end_date_time'])

print("Datos cargados correctamente")
print(f"Total de transacciones: {len(df):,}")
print(f"Periodo: {df['start_date_time'].min().date()} a {df['start_date_time'].max().date()}")
```

```python
# Verificar que los datos están correctos
df.head()
```

```python
# Gráfico de barras: Top 10 puertos más utilizados
plt.figure(figsize=(12, 6))

top_10 = df['evse_uid'].value_counts().head(10)
top_10.plot(kind='barh', color='skyblue', edgecolor='black')

plt.title('Top 10 Puertos de Carga Más Utilizados', fontsize=16, fontweight='bold')
plt.xlabel('Número de Transacciones', fontsize=12)
plt.ylabel('Puerto', fontsize=12)
plt.tight_layout()
plt.show()

print(f"\nEl puerto más usado es: {top_10.index[0]} con {top_10.values[0]:,} transacciones")
```

```python
# Gráfico de línea: Transacciones por mes
plt.figure(figsize=(12, 6))

trans_por_mes = df.groupby('mes_nombre').size()

# Ordenar por número de mes
orden_meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 
               'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
trans_por_mes = trans_por_mes.reindex([m for m in orden_meses if m in trans_por_mes.index])

trans_por_mes.plot(kind='line', marker='o', linewidth=2, markersize=8, color='blue')

plt.title('Evolución de Transacciones por Mes', fontsize=16, fontweight='bold')
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Número de Transacciones', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nMes con más transacciones: {trans_por_mes.idxmax()} ({trans_por_mes.max():,})")
print(f"Mes con menos transacciones: {trans_por_mes.idxmin()} ({trans_por_mes.min():,})")
```

```python
# Gráfico de barras: Transacciones por día de la semana
plt.figure(figsize=(10, 6))

orden_dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
trans_por_dia = df['dia_semana_esp'].value_counts().reindex(orden_dias)

trans_por_dia.plot(kind='bar', color='pink', edgecolor='black')

plt.title('Transacciones por Día de la Semana', fontsize=16, fontweight='bold')
plt.xlabel('Día', fontsize=12)
plt.ylabel('Número de Transacciones', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(f"\nDía más activo: {trans_por_dia.idxmax()} ({trans_por_dia.max():,} transacciones)")
print(f"Día menos activo: {trans_por_dia.idxmin()} ({trans_por_dia.min():,} transacciones)")
```

```python
# Gráfico circular: Distribución por periodo del día
plt.figure(figsize=(10, 8))

periodo_counts = df['periodo_dia'].value_counts()
colores = ['#FFD700', '#FF6347', '#4169E1', '#9370DB']

plt.pie(periodo_counts, labels=periodo_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=colores, textprops={'fontsize': 12})

plt.title('Distribución de Transacciones por Periodo del Día', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()

print("\nTransacciones por periodo:")
for periodo, cantidad in periodo_counts.items():
    print(f"  {periodo}: {cantidad:,} ({cantidad/len(df)*100:.1f}%)")
```

```python
# Histograma: Distribución de energía consumida
plt.figure(figsize=(12, 6))

plt.hist(df['energy_kwh'], bins=50, color='lightgreen', edgecolor='black', alpha=0.7)

plt.title('Distribución de Energía Consumida por Transacción', fontsize=16, fontweight='bold')
plt.xlabel('Energía (kWh)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.axvline(df['energy_kwh'].mean(), color='red', linestyle='--', linewidth=2, 
            label=f'Promedio: {df["energy_kwh"].mean():.2f} kWh')
plt.legend()
plt.tight_layout()
plt.show()

print(f"\nPromedio de energía por carga: {df['energy_kwh'].mean():.2f} kWh")
print(f"Mediana: {df['energy_kwh'].median():.2f} kWh")
```

```python
# Comparación: Días laborales vs Fin de semana
plt.figure(figsize=(10, 6))

tipo_dia = df['es_fin_de_semana'].map({False: 'Días Laborales', True: 'Fin de Semana'})
tipo_dia_counts = tipo_dia.value_counts()

tipo_dia_counts.plot(kind='bar', color=['steelblue', 'orange'], edgecolor='black')

plt.title('Transacciones: Días Laborales vs Fin de Semana', fontsize=16, fontweight='bold')
plt.xlabel('Tipo de Día', fontsize=12)
plt.ylabel('Número de Transacciones', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

laborales = (~df['es_fin_de_semana']).sum()
fin_semana = df['es_fin_de_semana'].sum()
print(f"\nDías laborales: {laborales:,} ({laborales/len(df)*100:.1f}%)")
print(f"Fin de semana: {fin_semana:,} ({fin_semana/len(df)*100:.1f}%)")
```

```python
# Gráfico de barras: Transacciones por hora del día
plt.figure(figsize=(14, 6))

trans_por_hora = df['hora'].value_counts().sort_index()
trans_por_hora.plot(kind='bar', color='purple', edgecolor='black', alpha=0.7)

plt.title('Transacciones por Hora del Día', fontsize=16, fontweight='bold')
plt.xlabel('Hora', fontsize=12)
plt.ylabel('Número de Transacciones', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nHora más activa: {trans_por_hora.idxmax()}:00 hrs ({trans_por_hora.max():,} transacciones)")
print(f"Hora menos activa: {trans_por_hora.idxmin()}:00 hrs ({trans_por_hora.min():,} transacciones)")
```

```python
print("="*80)
print("ANÁLISIS DE PUERTOS DE CARGA DE VEHÍCULOS ELÉCTRICOS")
print("="*80)

print("\n DATOS GENERALES:")
print(f"  • Total de transacciones analizadas: {len(df):,}")
print(f"  • Periodo de análisis: {df['start_date_time'].min().date()} a {df['start_date_time'].max().date()}")
print(f"  • Número de puertos diferentes: {df['evse_uid'].nunique()}")
print(f"  • Número de usuarios únicos: {df['user_id'].nunique()}")

print("\n CONSUMO DE ENERGÍA:")
print(f"  • Total energía consumida: {df['energy_kwh'].sum():,.2f} kWh")
print(f"  • Promedio por transacción: {df['energy_kwh'].mean():.2f} kWh")
print(f"  • Mínimo: {df['energy_kwh'].min():.2f} kWh")
print(f"  • Máximo: {df['energy_kwh'].max():.2f} kWh")

print("\n ANÁLISIS FINANCIERO:")
print(f"  • Ingresos totales: ${df['amount_transaction'].sum():,.0f}")
print(f"  • Ingreso promedio por transacción: ${df['amount_transaction'].mean():,.0f}")
print(f"  • Ingreso mínimo: ${df['amount_transaction'].min():,.0f}")
print(f"  • Ingreso máximo: ${df['amount_transaction'].max():,.0f}")

print("\n PUERTOS MÁS UTILIZADOS:")
top_5_puertos = df['evse_uid'].value_counts().head(5)
for i, (puerto, cantidad) in enumerate(top_5_puertos.items(), 1):
    porcentaje = (cantidad / len(df)) * 100
    print(f"  {i}. {puerto}: {cantidad:,} transacciones ({porcentaje:.1f}%)")

print("\n DISTRIBUCIÓN TEMPORAL:")
meses = df['mes_nombre'].value_counts()
print(f"  • Mes con más transacciones: {meses.idxmax()} ({meses.max():,})")
print(f"  • Mes con menos transacciones: {meses.idxmin()} ({meses.min():,})")

dias = df['dia_semana_esp'].value_counts()
print(f"  • Día más activo: {dias.idxmax()} ({dias.max():,} transacciones)")

horas = df['hora'].value_counts()
print(f"  • Hora pico: {horas.idxmax()}:00 hrs ({horas.max():,} transacciones)")

print("\n DISTRIBUCIÓN POR PERIODO DEL DÍA:")
periodos = df['periodo_dia'].value_counts()
for periodo, cantidad in periodos.items():
    porcentaje = (cantidad / len(df)) * 100
    print(f"  • {periodo}: {cantidad:,} transacciones ({porcentaje:.1f}%)")

print("\n DÍAS LABORALES VS FIN DE SEMANA:")
laborales = (~df['es_fin_de_semana']).sum()
fin_semana = df['es_fin_de_semana'].sum()
print(f"  • Días laborales: {laborales:,} ({laborales/len(df)*100:.1f}%)")
print(f"  • Fin de semana: {fin_semana:,} ({fin_semana/len(df)*100:.1f}%)")

print("\n" + "="*80)
print(" ANÁLISIS COMPLETADO")
print("="*80)
```
